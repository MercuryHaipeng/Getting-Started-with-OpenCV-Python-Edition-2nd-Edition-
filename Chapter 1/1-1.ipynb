#### 1.2 图像处理的基本操作

````python
from IPython.display import display, Image
import cv2
lena = cv2.imread("lena.bmp")
print(type(lena))
print(lena.shape)
print(lena) # 输出图像的像素值

_, img_encoded = cv2.imencode('.png', lena) # 转成 png 就能显示了
display(Image(data=img_encoded.tobytes())) # 内嵌展示
````

````python
if lena is not None:
	cv2.imshow("demo", lena) # 创建窗口demo，并显示图像
	cv2.waitKey(0)	# 无限等待，直到按下任意键
	cv2.destroyWindow("demo")	# 释放销毁指定窗口
````

````python
cv2.imshow("demo1", lena)
cv2.imshow("demo2", lena)
key = cv2.waitKey()
if key != -1:
    cv2.destroyAllWindows() # 释放销毁所有窗口
    print("触发了按键！")
````

````python
# 保存图像，保存成功，返回值为真，否则为假
r = cv2.imwrite("result.bmp", lena)
print(r)
````

````python
type(lena)
````

````python
# 对于类型检查错误，先转换为 UMat 再显示
# cv2.UMat() 全称是 Unified Matrix（统一矩阵）
# 它是 OpenCV 中一种特殊的数据结构, 自动利用 GPU（通过 OpenCL）或 CPU 进行加速
r = cv2.UMat(lena)
type(r)
````
