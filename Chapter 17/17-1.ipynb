#### 17.1 VideoCapture 类
- 构造函数 `cv2.VideoCapture()` 既支持视频文件读取，也支持从摄像头中读取视频，其语法格式为：
    - `捕获对象 = cv2.VidewCapture("摄像头ID")` 或 `捕获对象 = cv2.VidewCapture("文件名")`，其中
        - `捕获对象` 为返回值，是 `cv2.VideoCapture` 类的对象
        - `摄像头ID` 是摄像设备的ID编号，用数字“0”表示第1个摄像头（默认摄像头），“1”表示第2个摄像头，以此类推
        - `文件名` 可以是本地视频文件名、图像序列、视频流的URL等

- 函数 `retval = cv2.VideoCapture.isOpened()` 判断当前的摄像头是否初始化成功：
    - 如果成功，则返回值 `retval` 为 `Ture`
    - 如果不成功，则返回值 `retval` 为 `False`
- 函数 `retval = cv2.VideoCapture.open(param)` 用来打开摄像头或视频文件，其中：
    - `param` 可以为摄像头ID号 `index`，或者是文件名 `filename`
    - `retval` 为返回值，当摄像头或视频文件成功打开时，返回值为 `True`，否则为 `False`

- 函数 `retval, image = cv2.VideoCapture.read()` 用来捕获帧，其中：
    - `image` 返回捕获到的帧，如果没有帧被捕获，该值为空
    - `retval` 表示捕获是否成功，如果成功则该值为 `True`，否则为 `False`
- 在多相机环境中，尤其是没有硬件同步的情况下，需要用到 `cv2.VideoCapture.grab()` 和 `cv2.VideoCapture.retrieve()` 函数代替 `cv2.VideoCapture.read()`。通常情况下，为每个摄像头调用 `cv2.VideoCapture.grab()`，然后调用较慢的函数 `cv2.VideoCapture.retrieve()` 解码并从每个摄像头获取帧，使从不同摄像头检索到的帧在时间上更近
>```python
> success0 = cameraCapture0.grab()
> success1 = cameraCapture1.grab()
> if success0 and success1:
>     frame0 = cameraCapture0.retrieve().
>     frame1 = cameraCapture1.retrieve()
>```

- 函数 `None = cv2.VideoCapture.release()` 用来在释放资源时关闭摄像头，或者关闭文件资源

- 函数 `retval = cv2.VideoCapture.get(propId)` 用于获取 `cv2.VideoCapture` 类对象的属性，其中
    - `propId` 对应 `cv2.VideoCapture` 类对象的属性名称，或者是属性名称对应的数值
- 函数 `retval = cv2.VideoCapture.set(propId, value)` 用来设置 `cv2.VideoCapture` 类对象的属性，其中
    - `retval` 是返回值。属性设置不成功，返回 `False`；大多数情况下返回 `True`。注意返回 `True` 不一定代表绝对成功。
    - `propId` 对应 `cv2.VideoCapture` 类对象的属性名称，或者是属性名称对应的数值
    - `value` 对应属性 `propId` 要设置的新值

````python
# 捕获摄像头视频
import cv2

cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    cv2.imshow("frame", frame)
    c = cv2.waitKey(1)
    if c == 27:  # ESC键
        break
cap.release()
cv2.destroyAllWindows()
````

#### 17.2 VideoWirter 类
- 构造函数 `<VideoWriter object> = cv2.VideoWriter(filename, fourcc, fps, frameSize[, isColor])` 用来完成初始化，其中：
    - `filename` 指定输出目标视频的存放路径和文件名。如果指定的文件名已经存在，则会覆盖已经存在的文件
    - `fourcc` 表示视频编/解码类型。在OpenCV中用函数 `cv2.VideoWriter_fourcc()` 来指定视频编码格式，其参数由4个字符构成编/解码器的“4字标记”
    - `fps` 为帧速率，代表在1秒内出现的帧数
    - `frameSize` 为视频每一帧的宽度和高度
    - `isColor` 表示是否为彩色图像，注意当存储灰度图像时，必须保证要存储的帧是灰度形式的
- 函数 `None = cv2.VideoWriter.write(image)` 用于写入下一帧视频，其中：
    - `img` 是要写入的视频帧，通常要求彩色图像的格式为 BGR 模式
- 释放 `cv2.VideoWriter` 类对象的函数是 `None = cv2.VideoWriter.release()`

````python
# 使用 cv2.VideoWriter类保存捕获摄像头得到的彩色视频文件
import cv2
cap = cv2.VideoCapture(0)
w = cap.get(cv2.CAP_PROP_FRAME_WIDTH)   # 计算宽度 cv2.CAP_PROP_FRAME_WIDTH = 3
h = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)  # 计算高度 cv2.CAP_PROP_FRAME_HEIGHT = 4
size = (int(w), int(h))                    # 计算得到参数framesize
fourcc = cv2.VideoWriter_fourcc(*'MJPG')
out = cv2.VideoWriter('output1.avi',fourcc, 20, size, True)
print("size:", size)
# 参数frameSize使用获取的size
# 如果使用自定义大小，错误的大小值会导致存储失败
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == True:
        out.write(frame)
        cv2.imshow('frame', frame)
        if cv2.waitKey(1) == 27:
            break
    else:
        break
cap.release()
out.release()
cv2.destroyAllWindows()
````

````python
# 使用 cv2.VideoWriter类保存捕获摄像头得到视频文件的灰度形式
import cv2

cap = cv2.VideoCapture(0)
w = cap.get(cv2.CAP_PROP_FRAME_WIDTH)   # 计算宽度
h = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)      # 计算高度
size = (int(w), int(h))                 # 计算得到参数framesize
fourcc = cv2.VideoWriter_fourcc(*"MJPG")
out = cv2.VideoWriter("output2.avi", fourcc, 20, size, False)
print("size:", size)
# 使用获取的size,如果使用自定义大小，错误的大小值会导致存储失败
while cap.isOpened():
    ret, frame = cap.read()
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)  # 彩色转灰度
    if ret == True:
        out.write(frame)
        cv2.imshow("frame", frame)
        if cv2.waitKey(1) == 27:
            break
    else:
        break
cap.release()
out.release()
cv2.destroyAllWindows()
````

````python
import numpy as np
import cv2
cap = cv2.VideoCapture("Chapter 17/viptrain.avi")
# cap = cv2.VideoCapture("Chapter 17/output1_20260507.avi")
while(cap.isOpened()):
    ret, frame = cap.read()
    frame = cv2.Canny(frame, 100, 200)
    cv2.imshow('frame', frame)
    c = cv2.waitKey(1)
    if c == 27:       # ESC键
        break
cap.release()
cv2.destroyAllWindows()
````
